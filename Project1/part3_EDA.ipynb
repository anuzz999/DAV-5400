### 3.1 Summary Statistics:

We will calculate and display summary statistics like mean, median, and standard deviation for each attribute in the dataset.

Summary statistics provide a quick overview of the main aspects of the data, such as the central tendency, dispersion, and shape of the distribution. Here, we'll calculate:

-   Mean: The average value.
-   Median: The middle value that separates the higher half from the lower half.
-   Standard Deviation: A measure of the amount of variation or dispersion in the values.
-   Minimum and Maximum Values
-   Quartiles: Values that divide the data into four equal parts.





```python
from py2_optimized import ExploratoryDataAnalysis  

# Providing the file path to the class
eda = ExploratoryDataAnalysis("C:\\Users\\Asus\\Desktop\\Project1\\cleaned_data.csv")

# Calculating and displaying summary statistics
summary_statistics = eda.calculate_summary_statistics()
display(summary_statistics)  # Using display for a cleaner output in Jupyter notebooks


```

Here are the summary statistics for each numerical attribute in our dataset:

-   Count: Indicates the number of non-missing values for each column. All columns have 209,463 entries, which means there are no missing values left in these columns after cleaning.

-   Mean: Represents the average value in each column.

-   Std (Standard Deviation): Measures the dispersion or spread of the data values. A lower standard deviation indicates that the values are closer to the mean, while a higher standard deviation indicates that the values are spread out over a wider range.

-   Min: The smallest value in each column.

-   25% (1st Quartile): 25% of the data values are below this value.

-   50% (Median or 2nd Quartile): The middle value that separates the higher half from the lower half of the dataset.

-   75% (3rd Quartile): 75% of the data values are below this value.

-   Max: The largest value in each column.

### 3.3 Box Plot 

#### What are box plots?

-   Box plots are a visual representation that describe the distribution of a dataset.
-   They show the median, quartiles, and potential outliers in the dataset.

#### 3.3.1 Box of 'DTE' (Days to Expiry)



```python
# Calling the method to plot box plots for the 'DTE' column
eda.plot_boxplot(columns=['DTE'], titles=['Boxplot of DTE'], x_labels=['DTE'], y_labels=['Values'])
```

#### Observations from the 'DTE' Box Plot:

-   The median (the middle value) of 'DTE' is shown by the line inside the box.
-   The box represents the interquartile range (IQR), which is the range within which the middle 50% of data points fall.
-   The whiskers extend to the smallest and largest values within 1.5 times the IQR from the lower and upper quartiles, respectively.
-   Outliers are data points that fall outside this range and are displayed as individual points.

#### Interpretation:

-   Most options in the dataset have a 'DTE' value concentrated around the lower end, indicating that they are closer to expiry.
-   There are some options with a higher 'DTE', extending the whiskers and showing a few outliers. These options have expiry dates further in the future.

This visual insight allows us to understand the general expiry landscape of the options in our dataset.

### 3.3.2 Box of Greeks
Next, let's explore the Greeks in the dataset. Greeks are essential risk measures in options trading, and they help us understand the sensitivity of option prices to various factors.

We'll start with the Delta for both call and put options. Delta measures the sensitivity of an option's price to a $1 change in the price of the underlying asset.

1\. **Call Delta ('C_DELTA'):** Represents the change in the call option price relative to a change in the underlying asset price.

2\. **Put Delta ('P_DELTA'):** Represents the change in the put option price relative to a change in the underlying asset price.

We will create box plots for 'C_DELTA' and 'P_DELTA' to visualize the spread and identify any outliers.



```python
# Calling the method to plot box plots for the 'C_DELTA' and 'P_DELTA' columns
eda.plot_boxplot(columns=['C_DELTA', 'P_DELTA'], 
                titles=['Box Plot of Call Delta (C_DELTA)', 'Box Plot of Put Delta (P_DELTA)'], 
                x_labels=['C_DELTA', 'P_DELTA'], 
                y_labels=['Values', 'Values'])
```

### Box Plots of Call Delta ('C_DELTA') and Put Delta ('P_DELTA')

#### Observations:

-   Call Delta ('C_DELTA'):

    -   The box plot for Call Delta shows a wide spread of values, ranging from negative to positive. The median is around 0.5, which is expected for call options.
    -   There are several outliers on both ends of the distribution, indicating some options with delta values significantly away from the median.
-   Put Delta ('P_DELTA'):

    -   The distribution of Put Delta is concentrated towards the negative side, which is typical for put options, as they have negative deltas. The median is around -0.5.
    -   Similar to Call Delta, there are outliers in the Put Delta distribution as well, especially on the positive side.

#### Interpretation:

-   The Delta values' distribution and outliers provide insights into the options' moneyness in the dataset. Options with deltas close to 1 or -1 are deep in-the-money, while those close to 0 are out-of-the-money.

We will dive deeper into the Delta which will allow us to understand its characteristics and behavior better. We can explore the following aspects of Delta for both call and put options:

1.  Distribution Analysis:
    -   Explore the distribution of Delta values further, possibly categorizing them as in-the-money (ITM), at-the-money (ATM), and out-of-the-money (OTM).
2.  Delta vs. Days to Expiry (DTE):
    -   Analyze how Delta values change relative to the days remaining to expiry.
3.  Delta vs. Strike Distance:

We start by analyzing the distribution of Delta values, categorizing them into ITM, ATM, and OTM

Let's start by categorizing the Delta values into different money-ness categories:

- **In-the-money (ITM):**

  - Call options: Delta close to 1

  - Put options: Delta close to -1

- **At-the-money (ATM):**

  - Call and Put options: Delta close to 0

- **Out-of-the-money (OTM):**

  - Call options: Delta close to 0

  - Put options: Delta close to 0

We can create a histogram to visualize the distribution of Delta values and see how many options fall into each category.

-   The histograms give us an overview of how the time decay is distributed among the options, with most experiencing a mild to moderate rate of time decay.
